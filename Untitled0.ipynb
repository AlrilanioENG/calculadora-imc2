```python
def calcular_imc(peso, altura):
    imc = peso / (altura ** 2) # This line and the next should be indented
    return imc


def classificar_imc(imc):
    if imc < 18.5:
        return "Abaixo do peso"
    elif 18.5 <= imc < 24.9:
        return "Peso normal"
    elif 25 <= imc < 29.9:
        return "Sobrepeso"
    elif 30 <= imc < 34.9:
        return "Obesidade Grau 1"
    elif 35 <= imc < 39.9:
        return "Obesidade Grau 2"
    else:
        return "Obesidade Grau 3"

peso = float(input("Digite seu peso:")) # These lines need to call the float() function properly
altura = float(input("Digite sua altura: FORMATO EX 1.55"))

imc = calcular_imc(peso, altura)
categoria = classificar_imc(imc)


print(f"seu IMC é {imc:.2f}.Categoria: {categoria}")


```
